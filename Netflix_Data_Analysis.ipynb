## Data Cleaning

#### To keep only the text data in Title Column

```python
data = data[data['Title'].str.contains(r'[a-z]')]
```

### Convert the Seasons into minutes in Duration Column

```python
data["Duration"] = np.where(((data["Duration"].str.contains("Seasons")) | (data["Duration"].str.contains("Season"))), (data["Duration"].str.split().str[0].astype(int)*60*5).astype(str) + " min", data["Duration"])
```

```python
data["Duration"].head(2)
```

#### 1. To find the Duplicates in the data 

```python
data[data.duplicated()]
```

#### 2. To remove the Duplicates from the data 

```python
data.drop_duplicates(inplace = True)
```

```python
data
```

#### 3.To find the null values in the data

```python
data.isnull().sum()
```

#### 4. Representation of null values through heat-map

```python
sns.heatmap(data.isnull())
```

#### 5.Fill the null values

```python
data["Director"].fillna("No Director", inplace = True)
data["Cast"].fillna("No Cast", inplace = True)
data["Country"].fillna("No Country", inplace = True)
data["Release_Date"].fillna("No Release_Date", inplace = True)
data["Rating"].fillna("No Rating", inplace = True)
```

```python
data.isnull().sum()
```

```python
data["Release_Date"].replace("No Release_Date","12-Dec-99",inplace = True)
```
